```python
# probability of event A
favorable = 2
total = 10

P_A = favorable / total
print("P(A) =", P_A)
```

```python
P_A_and_B = 0.2
P_B = 0.5

P_A_given_B = P_A_and_B / P_B
print("P(A|B) =", P_A_given_B)
```

```python
# Bayes theorem
P_B_given_A = 0.8
P_A = 0.3
P_B = 0.5

P_A_given_B = (P_B_given_A * P_A) / P_B
print("P(A|B) =", P_A_given_B)
```

```python
import numpy as np

# previous belief
bel_prev = np.array([0.2, 0.5, 0.3])

# motion model
motion = np.array([
    [0.7, 0.2, 0.1],
    [0.2, 0.6, 0.2],
    [0.1, 0.2, 0.7]
])

# measurement model
measurement = np.array([0.9, 0.2, 0.5])

# prediction
bel_pred = motion.T @ bel_prev

# update
bel_new = measurement * bel_pred

# normalization
bel_new = bel_new / np.sum(bel_new)

print("Updated belief:", bel_new)
```

```python
import numpy as np

bel = np.array([0.3, 0.4, 0.3])

motion = np.array([
    [0.8, 0.1, 0.1],
    [0.2, 0.6, 0.2],
    [0.1, 0.2, 0.7]
])

measurements = [
    np.array([0.9, 0.3, 0.2]),
    np.array([0.2, 0.8, 0.3]),
    np.array([0.3, 0.4, 0.9])
]

for z in measurements:
    # prediction
    bel = motion.T @ bel
    
    # update
    bel = z * bel
    
    # normalize
    bel = bel / np.sum(bel)
    
    print("Belief:", bel)
```
